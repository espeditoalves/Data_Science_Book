### Propriedades da Função Logarítmica:
Em resumo, as transformações logarítmicas são úteis quando se deseja lidar com variação excessiva, linearizar relações não lineares ou estabilizar a variância em dados.

* A função logarítmica é a inversa da função exponencial. Ela se apresenta na forma geral: $$ y = \log_b(x) $$
* Aqui, (x) é o valor original e (y)  é o logaritmo desse valor na base (b). 
* A base (b) pode ser qualquer número maior que 1, mas as bases mais comuns são 10 (logaritmo comum) e  (e) (logaritmo natural).

### Compressão e Expansão de Intervalos:

A função logarítmica comprime o intervalo de grandes números e expande o intervalo de pequenos números. Por exemplo, um intervalo entre 100 e 1000 é comprimido para um intervalo entre 2 e 3 quando aplicamos o logaritmo na base 10.

### Visualização de Padrões:

As transformações logarítmicas são úteis para facilitar a visualização de padrões nos dados. Elas transformam relações exponenciais em relações mais compreensíveis.

### Aplicações Práticas:

Na ciência de dados, usamos transformações logarítmicas quando:

- A variável tem uma distribuição assimétrica.
- A variância dos dados é grande.

A transformação logarítmica é definida como: $$ y = \log(x) $$


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
diamonds = sns.load_dataset("diamonds")
diamonds.head()
```

```python
diamonds.hist(figsize=(16, 16))
```

```python
diamonds['price_log'] = np.log(diamonds['price'])
diamonds['carat_log'] = np.log(diamonds['carat'])
```

```python
f, axs = plt.subplots(2, 2, figsize=(8, 4), gridspec_kw=dict(width_ratios=[8, 6]))
# Transforme axs em uma lista unidimensional
axs = axs.flatten()
sns.histplot(data=diamonds, x="price", ax=axs[0])
sns.histplot(data=diamonds, x="price_log", ax=axs[1])
sns.histplot(data=diamonds, x="carat", ax=axs[2])
sns.histplot(data=diamonds, x="carat_log", ax=axs[3])
f.tight_layout()


# Calcule a assimetria
skewness_price = diamonds["price"].skew()
skewness_price_log = diamonds["price_log"].skew()
skewness_carat = diamonds["carat"].skew()
skewness_carat_log = diamonds["carat_log"].skew()
# Atualize os títulos dos gráficos
axs[0].set_title(f"Assimetria (price): {skewness_price:.2f}")
axs[1].set_title(f"Assimetria (price_log): {skewness_price_log:.2f}")
axs[2].set_title(f"Assimetria (carat): {skewness_carat:.2f}")
axs[3].set_title(f"Assimetria (carat_log): {skewness_carat_log:.2f}")

plt.show()
```

```python
df = pd.DataFrame([1, 10, 20, 25, 30, 40, 100, 110, 200, 300, 500, 800, 1000, 1050, 2000, 10000, 500000], columns=['price'])
df['price_log_e'] = np.log(df['price'])
df['price_log10'] = np.log10(df['price'])
df['price_log2'] = np.log2(df['price'])
# Exibindo as primeiras linhas do DataFrame
df
```

```python
f, axs = plt.subplots(2, 2, figsize=(8, 6), gridspec_kw=dict(width_ratios=[8, 6]))
# Transforme axs em uma lista unidimensional
axs = axs.flatten()
sns.histplot(data=df, x="price", ax=axs[0])
sns.histplot(data=df, x="price_log_e", ax=axs[1])
sns.histplot(data=df, x="price_log10", ax=axs[2])
sns.histplot(data=df, x="price_log2", ax=axs[3])
f.tight_layout()


# Calcule a assimetria
skewness_price = df["price"].skew()
skewness_price_log = df["price_log_e"].skew()
skewness_carat = df["price_log10"].skew()
skewness_carat_log = df["price_log2"].skew()
# Atualize os títulos dos gráficos
axs[0].set_title(f"Assimetria (price): {skewness_price:.2f}")
axs[1].set_title(f"Assimetria (price_log_e): {skewness_price_log:.2f}")
axs[2].set_title(f"Assimetria (price_log10): {skewness_carat:.2f}")
axs[3].set_title(f"Assimetria (price_log2): {skewness_carat_log:.2f}")

plt.show()
```
